```python
def f(x):
  return ((a)+(b*x)+(c*x*x))

xl = []
yl = []
ff = []
a = 1
b = 2
c = 3
y = [0.05,0.25,0.75,0.95]

for i in range(0,11):
  xl.append(i/10)
  yl.append(f(xl[i]))
  ff.append([xl[i], yl[i]])

def lagrange(f,x,n):
  result = 0.0
  for i in range(n): 
    final = f[i][1]
    for j in range(n): 
      if j!= i: 
        final = ((x - f[j][0]) / (f[i][0] - f[j][0])) * final
      else:
        continue
    result += final
  return result 

count=0
for k in range(len(y)):
  if abs(f(y[k]) - lagrange(ff, y[k], len(yl)-1))<10**-10:
    # print(f"{f(y[k])} ~ {lagrange(ff, y[k], len(yl)-1)}")
    count = count+1
  if count == len(y):
    print("Values match")
```
